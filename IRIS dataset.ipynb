# CDF

```python
  #plot CDF  of petal_length
np.seterr(invalid='ignore')    
counts,bin_edges=np.histogram(iris_setosa['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)

#compute CDF
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)



# for iris_virginica
counts,bin_edges=np.histogram(iris_virginica['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)

#compute CDF
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


#for iris_versicolor
counts,bin_edges=np.histogram(iris_versicolor['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)

#compute CDF
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


plt.show()
```

# Mean,  Variance and Std Dev

```python
# Mean, varience and std dev
```

```python
#mean, varience and std dev can be easily corrupted by the use a small number of outliers
```

```python
print("Means")
print("The mean for setosa is "+str(np.mean(iris_setosa["petal_length"])))
print("The mean for virginica is "+str(np.mean(iris_virginica["petal_length"])))
print("The mean for versicolor is "+str(np.mean(iris_versicolor["petal_length"])))

# lets consider outlier to get to know more about mean 
print("The outlier mean for setosa is "+str(np.mean(np.append(iris_setosa["petal_length"],50))))

print("\n")
print("Standard Deviation:\n")
print("The mean for setosa is "+str(np.std(iris_setosa["petal_length"])))
print("The mean for virginica is "+str(np.std(iris_virginica["petal_length"])))
print("The mean for versicolor is "+str(np.std(iris_versicolor["petal_length"])))
```

# Median, Percentile,Quantile,IQR,MAD

```python
print("Medians\n")
print(np.median(iris_setosa["petal_length"]))

print(np.median(iris_virginica["petal_length"]))
print(np.median(iris_versicolor["petal_length"]))
print("\n")
print("The median of setosa petal length with an outlier 50 is")
print(np.median(np.append(iris_setosa["petal_length"],50)))

#Median is not effected much by a small number of outliers.
print("Quantiles")
print(np.percentile(iris_setosa["petal_length"],np.arange(0,100,25)))
print(np.percentile(iris_virginica["petal_length"],np.arange(0,100,25)))
print(np.percentile(iris_versicolor["petal_length"],np.arange(0,100,25)))

print("90 percentile\n")
print("Quantiles")
print(np.percentile(iris_setosa["petal_length"],90))
print(np.percentile(iris_virginica["petal_length"],90))
print(np.percentile(iris_versicolor["petal_length"],90))

from statsmodels import robust
print("Median Absolute Deviation\n")
print(robust.mad(iris_setosa["petal_length"]))
print(robust.mad(iris_virginica["petal_length"]))
print(robust.mad(iris_versicolor["petal_length"]))

#IQR=75%ile-25%ile = Inter quartile range
```
